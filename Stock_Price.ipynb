```python
df = yf.download('AAPL', start='2014-01-01', end=datetime.now())
df
```

```python
plt.figure(figsize=(14,6))
plt.title('Close Price History')
plt.plot(df['Close'])
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)
plt.show()
```

```python
data = df.filter(['Close'])
len(data)
dataset = data.values
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

scaled_data
```

```python
training_data_len = int(np.ceil( len(dataset) * .95 ))
training_data_len
```

```python
train_data = scaled_data[0:int(training_data_len), :]
# Split the data into x_train and y_train data sets
x_train = []
y_train = []

for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])
    if i == 60:
        print(x_train)
        print(y_train)
        
# Convert the x_train and y_train to numpy arrays since NumPy arrays are the standard input format for most deep learning libraries.
x_train, y_train = np.array(x_train), np.array(y_train)

# Reshape the data because RNNs, including LSTMs, expect input data in 3D format
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Input
```

```python
model = Sequential()

# Define the input layer explicitly
model.add(Input(shape=(x_train.shape[1], 1)))

# Add the LSTM layers
model.add(LSTM(128, return_sequences=True))
model.add(LSTM(64, return_sequences=False))

# Add the Dense layers
model.add(Dense(25))
model.add(Dense(1))
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')
# Train the model
model.fit(x_train, y_train, batch_size=1, epochs=1)
```

```python
training_data_len - 60
```

```python
# Create the testing data set
# Create a new array containing scaled values from index 1543 to 2002 
test_data = scaled_data[training_data_len - 60:, :]
len(test_data)
```

```python
x_test = []
y_test = dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1 ))
```

```python
x_test
```

```python
# Get the models predicted price values 
predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

# Get the root mean squared error (RMSE)
rmse = np.sqrt(np.mean(((predictions - y_test) ** 2)))
rmse
```

```python
# Assuming 'data' is a DataFrame containing the stock prices
train = data[:training_data_len]
valid = data[training_data_len:].copy()  # Create a copy of the validation data

# Add predictions to the validation DataFrame
valid['Predictions'] = predictions

# Visualize the data
plt.figure(figsize=(16, 6))
plt.title('Model')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Close Price USD ($)', fontsize=18)

# Plot training data
plt.plot(train['Close'], label='Train')
# Plot actual and predicted validation data
plt.plot(valid['Close'], label='Val')
plt.plot(valid['Predictions'], label='Predictions')

plt.legend(loc='lower right')
plt.show()
```

```python
def get_stock_price(stock_symbol, target_date):
    # Define the date range for fetching historical data
    end_date = pd.to_datetime(target_date)
    start_date = end_date - pd.DateOffset(years=1)  # Fetch data for the past year or adjust as needed

    # Fetch historical data from yfinance
    historical_data = yf.download(stock_symbol, start=start_date, end=end_date + pd.DateOffset(days=1))

    # Ensure the data is sorted by date
    historical_data.index = pd.to_datetime(historical_data.index)
    historical_data.sort_index(inplace=True)

    # Check if the target date is in the DataFrame index
    if target_date in historical_data.index:
        actual_price = historical_data.loc[target_date]['Close']
        return actual_price
    else:
        return None
```

```python
from datetime import datetime, timedelta
def calculate_start_date(prediction_date_str, window_days):
    """
    Calculate the start date given a prediction date and a time window.

    :param prediction_date_str: The date for which prediction is to be made (format 'YYYY-MM-DD').
    :param window_days: Number of days of historical data required.
    :return: Start date for fetching historical data.
    """
    # Convert prediction_date_str to datetime object
    prediction_date = datetime.strptime(prediction_date_str, '%Y-%m-%d')
    
    # Calculate the start date by subtracting the window_days from the prediction_date
    start_date = prediction_date - timedelta(days=window_days)
    
    # Return the start date in string format
    return start_date.strftime('%Y-%m-%d')
```

```python
# Define the stock symbol and prediction date
stock_symbol = 'AAPL'
prediction_date = '2024-07-25'
start_date = calculate_start_date(prediction_date, 60)
start_date
```
